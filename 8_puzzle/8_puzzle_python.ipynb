```python
from collections import deque
import copy
import sys

def inversionCount(arr):
  count = 0
  for i in range(len(arr)):
    for j in range(i + 1, len(arr)):
      if arr[i] > arr[j]:
        count += 1
  return count

def isDestinationReached(initial, final):
  for i in range(3):
    for j in range(3):
      if initial[i][j] != final[i][j]:
        return False
  return True

def notPresent(initial, vis):
  for i in range(len(vis)):
    isPresent = True
    for j in range(3):
      for k in range(3):
        if vis[i][j][k] != initial[j][k]:
          isPresent = False
    if isPresent:
      return False

  return True

def up(initial, curr_pos):
  if curr_pos[0] == 0:
    return False

  initial = copy.deepcopy(initial)
  initial[curr_pos[0]][curr_pos[1]] = initial[curr_pos[0] - 1][curr_pos[1]]
  initial[curr_pos[0] - 1][curr_pos[1]] = 0

  return [[curr_pos[0] - 1, curr_pos[1]], initial]

def down(initial, curr_pos):
  if curr_pos[0] == 2:
    return False

  initial = copy.deepcopy(initial)
  initial[curr_pos[0]][curr_pos[1]] = initial[curr_pos[0] + 1][curr_pos[1]]
  initial[curr_pos[0] + 1][curr_pos[1]] = 0
  return[[curr_pos[0] + 1, curr_pos[1]], initial]

def left(initial, curr_pos):
  if curr_pos[1] == 0:
    return False

  initial = copy.deepcopy(initial)
  initial[curr_pos[0]][curr_pos[1]] = initial[curr_pos[0]][curr_pos[1] - 1]
  initial[curr_pos[0]][curr_pos[1] - 1] = 0
  return [[curr_pos[0], curr_pos[1] - 1], initial]

def right(initial, curr_pos):
  if curr_pos[1] == 2:
    return False

  initial = copy.deepcopy(initial)
  initial[curr_pos[0]][curr_pos[1]] = initial[curr_pos[0]][curr_pos[1] + 1]
  initial[curr_pos[0]][curr_pos[1] + 1] = 0
  return [[curr_pos[0], curr_pos[1] + 1], initial]


def solve(initial, empty_pos, final, vis):
  queue = deque()
  queue.append([empty_pos, initial, [initial]])
  vis.append(initial)

  while queue:

    arr = queue.popleft()
    current_pos = arr[0]
    initial = arr[1]

    if isDestinationReached(initial, final):
      return arr[2]


    isrichable = up(initial, current_pos)
    if isrichable and notPresent(isrichable[1], vis):
      vis.append(isrichable[1])
      path = copy.deepcopy(arr[2])
      path.append(isrichable[1])
      isrichable.append(path)
      queue.append(isrichable)

    isrichable = down(initial, current_pos)
    if isrichable and notPresent(isrichable[1], vis):
      vis.append(isrichable[1])
      path = copy.deepcopy(arr[2])
      path.append(isrichable[1])
      isrichable.append(path)
      queue.append(isrichable)

    isrichable = left(initial, current_pos)
    if isrichable and notPresent(isrichable[1], vis):
      vis.append(isrichable[1])
      path = copy.deepcopy(arr[2])
      path.append(isrichable[1])
      isrichable.append(path)
      queue.append(isrichable)

    isrichable = right(initial, current_pos)
    if isrichable and notPresent(isrichable[1], vis):
      vis.append(isrichable[1])
      path = copy.deepcopy(arr[2])
      path.append(isrichable[1])
      isrichable.append(path)
      queue.append(isrichable)

  return None



print("Enter the input Matrix : ")
initial = [ [ 1, 8, 2 ],
            [ 0, 4, 3 ],
            [ 7, 6, 5 ] ]

# for i in range(3):
#   for j in range(3):
#     initial[i][j] = int(input(f"Enter the value for location [{i}, {j}] : "))

print("Initial configuration : ")
for i in initial:
  print(i)
print()
# Solvable Final configuration
# Value 0 is used for empty space

print("Enter the output Matrix : ")

final = [ [ 1, 2, 3 ],
          [ 4, 5, 6 ],
          [ 7, 8, 0 ] ]

# for i in range(3):
#   for j in range(3):
#     final[i][j] = int(input(f"Enter the value for location [{i}, {j}] : "))

print("Final configuration : ")
for i in final:
  print(i)
print()
# Blank tile coordinates in
# initial configuration
empty_pos = []

for i in range(3):
  for j in range(3):
    if initial[i][j] == 0:
      empty_pos = [i, j]
      break

vis = []

flatten_initial = []
for i in initial:
  flatten_initial.extend(i)

flatten_initial.remove(0)

if (inversionCount(flatten_initial) % 2) == 1:
  print("Not Solvable")
else:
  print("Solvable")
  path = solve(initial, empty_pos, final, vis)

  print("path to the solution : ")
  for i in path:
    for j in i:
      print(j)
    print()
```
